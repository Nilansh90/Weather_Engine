## CREATING A WILL RAIN OR NOT PREDICT MODEL

```python
import pandas as pd
import numpy as np
import lightgbm as lgb
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# --- STEP 1: Feature Engineering for Wind Speed (Lags & Rolling Windows) ---
# Combine temporarily to ensure chronological order per city for historical features
Unified_Train_Set['split'] = 'train'
Unified_Val_Set['split'] = 'val'
Unified_Test_Set['split'] = 'test'

full_df = pd.concat([Unified_Train_Set, Unified_Val_Set, Unified_Test_Set], ignore_index=True)

# Ensure sorted order for accurate rolling windows per city group
full_df = full_df.sort_values(by=['city_id', 'year', 'day']).reset_index(drop=True)

# Generate lag and rolling features from historical wind speed max
full_df['wind_lag_1'] = full_df.groupby('city_id')['wind_speed_max'].shift(1)
full_df['wind_lag_2'] = full_df.groupby('city_id')['wind_speed_max'].shift(2)
full_df['wind_roll_mean_3'] = full_df.groupby('city_id')['wind_speed_max'].transform(lambda x: x.shift(1).rolling(3).mean())
full_df['wind_roll_mean_7'] = full_df.groupby('city_id')['wind_speed_max'].transform(lambda x: x.shift(1).rolling(7).mean())

# Handle any initial edge cases due to the shifting windows by backward filling
historical_wind_cols = ['wind_lag_1', 'wind_lag_2', 'wind_roll_mean_3', 'wind_roll_mean_7']
for col in historical_wind_cols:
    full_df[col] = full_df.groupby('city_id')[col].bfill()

# Split back to clean Train, Val, and Test subsets
Train_Final = full_df[full_df['split'] == 'train'].copy()
Val_Final = full_df[full_df['split'] == 'val'].copy()
Test_Final = full_df[full_df['split'] == 'test'].copy()

# --- STEP 2: Use Proxy Model Forecasts for Validation and Test ---
base_features = ['city_id', 'year', 'day', 'dayofweek', 'month_sin', 'month_cos']

for df in [Val_Final, Test_Final]:
    # Replace true values with expected values from preceding models
    df['temp_max'] = temp_max_model.predict(df[base_features])
    
    chained_features = base_features + ['temp_max']
    df['temp_min'] = temp_min_model.predict(df[chained_features])
```

```python
# --- STEP 3: Setup Classification Labels and Feature Space ---
# Define target label (1 if it rained, 0 if dry)
for df in [Train_Final, Val_Final, Test_Final]:
    df['will_precip'] = (df['precip_sum'] > 0).astype(int)

# Include your explicit base parameters + your new historical wind parameters
features = [
    'city_id', 'temp_max', 'temp_min', 'year', 'day', 'dayofweek', 'month_sin', 'month_cos',
    'wind_lag_1', 'wind_lag_2', 'wind_roll_mean_3', 'wind_roll_mean_7'
]

# Build training segments
X_train, y_train = Train_Final[features], Train_Final['will_precip']
X_val, y_val = Val_Final[features], Val_Final['will_precip']
X_test, y_test = Test_Final[features], Test_Final['will_precip']

# --- STEP 4: Instantiate & Fit Classifier ---
Predict_precip = lgb.LGBMClassifier(
    objective='binary',
    random_state=42,
    n_estimators=1000,
    learning_rate=0.03,
    scale_pos_weight=2.0,
    colsample_bytree=0.8,
    num_leaves=31,
    reg_alpha=0.1,              
    reg_lambda=0.5,             
    class_weight='balanced',
    n_jobs=-1
    )

Predict_precip.fit(
    X_train, y_train,
    eval_set=[(X_val, y_val)],
    callbacks=[lgb.early_stopping(stopping_rounds=15, verbose=False)]
)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# --- STEP 5: Metric Evaluation Framework ---
def evaluate_dataset(name, X, y_true):
    # Use your exact model name to predict classes and probabilities
    preds = Predict_precip.predict(X)
    probs = Predict_precip.predict_proba(X)[:, 1]
    
    print(f"--- Performance Evaluation: {name.upper()} SET ---")
    print(f"Accuracy:  {accuracy_score(y_true, preds):.4f}")
    print(f"Precision: {precision_score(y_true, preds, zero_division=0):.4f}")
    print(f"Recall:    {recall_score(y_true, preds):.4f}")
    print(f"F1-Score:  {f1_score(y_true, preds):.4f}")
    print(f"ROC-AUC:   {roc_auc_score(y_true, probs):.4f}\n")

# Run assessment metrics across all data steps
evaluate_dataset("Train (Ground Truth Grounding)", X_train, y_train)
evaluate_dataset("Validation (Model Predictions Input)", X_val, y_val)
evaluate_dataset("Test (Model Predictions Input)", X_test, y_test)
```

## Reducing overfitting(mild)

```python
import lightgbm as lgb

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)

# ==========================================================
# 1. Feature set
# ==========================================================
base_model_features = [
    'city_id',
    'temp_max',
    'temp_min',
    'year',
    'day',
    'dayofweek',
    'month_sin',
    'month_cos',
    'wind_lag_1',
    'wind_lag_2',
    'wind_roll_mean_3',
    'wind_roll_mean_7'
]

# ==========================================================
# 2. Build datasets
# ==========================================================
X_train = Train_Final[base_model_features]
y_train = Train_Final['will_precip']

X_val = Val_Final[base_model_features]
y_val = Val_Final['will_precip']

X_test = Test_Final[base_model_features]
y_test = Test_Final['will_precip']


# ==========================================================
# 3. Create NEW LightGBM model
# ==========================================================
Predict_precip = lgb.LGBMClassifier(

    objective='binary',
    boosting_type='gbdt',

    learning_rate=0.02,
    n_estimators=3000,

    num_leaves=7,
    max_depth=4,

    min_child_samples=75,

    subsample=0.7,
    subsample_freq=1,

    colsample_bytree=0.7,

    reg_alpha=2.0,
    reg_lambda=5.0,

    scale_pos_weight=2.0,

    random_state=42,
    n_jobs=-1
)

print(Predict_precip)


# ==========================================================
# 4. Train model
# ==========================================================
Predict_precip.fit(

    X_train,
    y_train,

    eval_set=[(X_val, y_val)],

    callbacks=[
        lgb.early_stopping(
            stopping_rounds=50,
            verbose=False
        )
    ]
)


print("Best iteration:",
      Predict_precip.best_iteration_)


# ==========================================================
# 5. Evaluation function
# ==========================================================
def evaluate_model(name, X, y):

    pred = Predict_precip.predict(X)

    prob = Predict_precip.predict_proba(X)[:,1]


    print(f"\n--- {name.upper()} ---")

    print(f"Accuracy : {accuracy_score(y,pred):.4f}")

    print(f"Precision: {precision_score(y,pred,zero_division=0):.4f}")

    print(f"Recall   : {recall_score(y,pred):.4f}")

    print(f"F1-Score : {f1_score(y,pred):.4f}")

    print(f"ROC-AUC  : {roc_auc_score(y,prob):.4f}")


# ==========================================================
# 6. Evaluate
# ==========================================================
evaluate_model("Train", X_train, y_train)

evaluate_model("Validation", X_val, y_val)

evaluate_model("Test", X_test, y_test)
```

## CODE FOR PREDICTING PRECIP AMOUNT

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# 1. Define feature lists correctly
# These must match the columns that exist in Train_Final, Val_Final, and Test_Final
# (If you don't have these, you must re-run the cell where you created full_df)
reg_features = [
    'year', 'month_sin', 'month_cos', 'day', 'dayofweek', 'city_id', 
    'rain_prob', 'wind_speed_max', 'temp_max', 'temp_min',
    'wind_lag_1', 'wind_lag_2', 'wind_roll_mean_3', 'wind_roll_mean_7'
]

# 2. Add 'rain_prob' to the Final DataFrames
# Predict_precip was trained on base_model_features, so we use that to get probas
for df in [Train_Final, Val_Final, Test_Final]:
    df['rain_prob'] = Predict_precip.predict_proba(df[base_model_features])[:, 1]

# 3. Prepare Matrices
# Use the DataFrames that contain the lag/rolling features
X_train_reg = Train_Final[reg_features]
y_train_reg = Train_Final['precip_sum']

X_val_reg = Val_Final[reg_features]
y_val_reg = Val_Final['precip_sum']

X_test_reg = Test_Final[reg_features]
y_test_reg = Test_Final['precip_sum']

# 4. Initialize and Train the Regressor
# 1. Initialize the Regressor with Tweedie objective
# Tweedie power 1.5 is a common starting point for precipitation
lgb_reg = lgb.LGBMRegressor(
    objective='tweedie',
    tweedie_variance_power=1.5,  
    n_estimators=1000,
    learning_rate=0.05,
    random_state=42,
    n_jobs=-1
)

# 2. Fit the model
lgb_reg.fit(
    X_train_reg, y_train_reg,
    eval_set=[(X_val_reg, y_val_reg)],
    callbacks=[lgb.early_stopping(stopping_rounds=50, verbose=False)]
)

# 3. Evaluation function (no changes needed)
def evaluate_regressor(name, X, y_true):
    preds = lgb_reg.predict(X)
    mae = mean_absolute_error(y_true, preds)
    rmse = np.sqrt(mean_squared_error(y_true, preds))
    r2 = r2_score(y_true, preds)
    
    print(f"\n--- Performance (Tweedie): {name.upper()} SET ---")
    print(f"MAE:  {mae:.4f}")
    print(f"RMSE: {rmse:.4f}")
    print(f"R²:   {r2:.4f}")

evaluate_regressor("Train", X_train_reg, y_train_reg)
evaluate_regressor("Validation", X_val_reg, y_val_reg)
evaluate_regressor("Test", X_test_reg, y_test_reg)

lgb_reg.fit(
    X_train_reg, y_train_reg,
    eval_set=[(X_val_reg, y_val_reg)],
    callbacks=[lgb.early_stopping(stopping_rounds=50, verbose=False)]
)

# 5. Evaluation Function
def evaluate_regressor(name, X, y_true):
    preds = lgb_reg.predict(X)
    mae = mean_absolute_error(y_true, preds)
    rmse = np.sqrt(mean_squared_error(y_true, preds))
    r2 = r2_score(y_true, preds)
    
    print(f"\n--- Performance: {name.upper()} SET ---")
    print(f"MAE:  {mae:.4f}")
    print(f"RMSE: {rmse:.4f}")
    print(f"R²:   {r2:.4f}")

evaluate_regressor("Train", X_train_reg, y_train_reg)
evaluate_regressor("Validation", X_val_reg, y_val_reg)
evaluate_regressor("Test", X_test_reg, y_test_reg)
```

```python
import pandas as pd
import numpy as np
import lightgbm as lgb
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# 1. Re-calculate Features on the full dataset to include new Precipitation features
# (Assuming Unified_Train_Set, Unified_Val_Set, Unified_Test_Set exist)
full_df = pd.concat([Unified_Train_Set, Unified_Val_Set, Unified_Test_Set], ignore_index=True)
full_df = full_df.sort_values(by=['city_id', 'year', 'day']).reset_index(drop=True)

# --- Wind Features ---
full_df['wind_lag_1'] = full_df.groupby('city_id')['wind_speed_max'].shift(1)
full_df['wind_lag_2'] = full_df.groupby('city_id')['wind_speed_max'].shift(2)
full_df['wind_roll_mean_3'] = full_df.groupby('city_id')['wind_speed_max'].transform(lambda x: x.shift(1).rolling(3).mean())
full_df['wind_roll_mean_7'] = full_df.groupby('city_id')['wind_speed_max'].transform(lambda x: x.shift(1).rolling(7).mean())

# --- NEW: Precipitation Historical Features ---
full_df['precip_lag_1'] = full_df.groupby('city_id')['precip_sum'].shift(1)
full_df['precip_lag_2'] = full_df.groupby('city_id')['precip_sum'].shift(2)
full_df['precip_lag_3'] = full_df.groupby('city_id')['precip_sum'].shift(3)
full_df['precip_roll_mean_3'] = full_df.groupby('city_id')['precip_sum'].transform(lambda x: x.shift(1).rolling(3).mean())
full_df['precip_roll_mean_7'] = full_df.groupby('city_id')['precip_sum'].transform(lambda x: x.shift(1).rolling(7).mean())
full_df['precip_roll_max_7'] = full_df.groupby('city_id')['precip_sum'].transform(lambda x: x.shift(1).rolling(7).max())

# Handle initial NaNs
history_cols = [
    'wind_lag_1', 'wind_lag_2', 'wind_roll_mean_3', 'wind_roll_mean_7',
    'precip_lag_1', 'precip_lag_2', 'precip_lag_3', 
    'precip_roll_mean_3', 'precip_roll_mean_7', 'precip_roll_max_7'
]
for col in history_cols:
    full_df[col] = full_df.groupby('city_id')[col].bfill()

# Split back to clean sets
Train_Final = full_df[full_df['split'] == 'train'].copy()
Val_Final = full_df[full_df['split'] == 'val'].copy()
Test_Final = full_df[full_df['split'] == 'test'].copy()

# 2. Add 'rain_prob' using the Classifier
# Ensure base_model_features is defined as used in your Classifier
for df in [Train_Final, Val_Final, Test_Final]:
    df['rain_prob'] = Predict_precip.predict_proba(df[base_model_features])[:, 1]

# 3. Define Regressor Features
reg_features = [
    'year', 'month_sin', 'month_cos', 'day', 'dayofweek', 'city_id', 
    'rain_prob', 'wind_speed_max', 'temp_max', 'temp_min',
    'wind_lag_1', 'wind_lag_2', 'wind_roll_mean_3', 'wind_roll_mean_7',
    'precip_lag_1', 'precip_lag_2', 'precip_lag_3', 
    'precip_roll_mean_3', 'precip_roll_mean_7', 'precip_roll_max_7'
]

# 4. Train Regressor
X_train_reg, y_train_reg = Train_Final[reg_features], Train_Final['precip_sum']
X_val_reg, y_val_reg = Val_Final[reg_features], Val_Final['precip_sum']
X_test_reg, y_test_reg = Test_Final[reg_features], Test_Final['precip_sum']

lgb_reg = lgb.LGBMRegressor(
    objective='tweedie',
    tweedie_variance_power=1.5,
    n_estimators=1000,
    learning_rate=0.05,
    random_state=42
)

lgb_reg.fit(
    X_train_reg, y_train_reg,
    eval_set=[(X_val_reg, y_val_reg)],
    callbacks=[lgb.early_stopping(50, verbose=False)]
)

# 5. Evaluate
def evaluate_regressor(name, X, y_true):
    preds = lgb_reg.predict(X)
    print(f"\n--- Performance: {name.upper()} SET ---")
    print(f"MAE:  {mean_absolute_error(y_true, preds):.4f}")
    print(f"RMSE: {np.sqrt(mean_squared_error(y_true, preds)):.4f}")
    print(f"R²:   {r2_score(y_true, preds):.4f}")

evaluate_regressor("Train", X_train_reg, y_train_reg)
evaluate_regressor("Validation", X_val_reg, y_val_reg)
evaluate_regressor("Test", X_test_reg, y_test_reg)
```

```python

import itertools

# 1. Hyperparameter Grid
powers = [1.1, 1.3, 1.5, 1.7]
leaves = [7, 15, 31]
child = [30, 50, 75]

best_score = float('inf')
best_params = {}
best_model = None

# Log-transform target
y_train_log = np.log1p(y_train_reg)
y_val_log = np.log1p(y_val_reg)

# 2. Grid Search
for p, l, c in itertools.product(powers, leaves, child):
    model = lgb.LGBMRegressor(
        objective='tweedie',
        tweedie_variance_power=p,
        num_leaves=l,
        min_child_samples=c,
        n_estimators=1000,
        learning_rate=0.05,
        random_state=42,
        n_jobs=-1
    )
    
    model.fit(
        X_train_reg, y_train_log,
        eval_set=[(X_val_reg, y_val_log)],
        callbacks=[lgb.early_stopping(50, verbose=False)]
    )
    
    # Predict and invert log transform
    val_preds = np.expm1(model.predict(X_val_reg))
    rmse = np.sqrt(mean_squared_error(y_val_reg, val_preds))
    
    if rmse < best_score:
        best_score = rmse
        best_params = {'power': p, 'leaves': l, 'child': c}
        best_model = model

print(f"Best Params: {best_params}")

# 3. Final Evaluation with Best Model
def evaluate_final(name, X, y_true):
    preds = np.expm1(best_model.predict(X))
    mae = mean_absolute_error(y_true, preds)
    rmse = np.sqrt(mean_squared_error(y_true, preds))
    r2 = r2_score(y_true, preds)
    
    print(f"\n--- Performance (Log-Tweedie): {name.upper()} SET ---")
    print(f"MAE:  {mae:.4f}")
    print(f"RMSE: {rmse:.4f}")
    print(f"R²:   {r2:.4f}")

evaluate_final("Train", X_train_reg, y_train_reg)
evaluate_final("Validation", X_val_reg, y_val_reg)
evaluate_final("Test", X_test_reg, y_test_reg)
```
