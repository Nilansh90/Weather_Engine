```python
DEL_TRAIN=pd.read_csv("../data/processed/Delhi/train.csv")
BLR_TRAIN = pd.read_csv("../data/processed/Bangalore/train.csv")
MUM_TRAIN = pd.read_csv("../data/processed/Mumbai/train.csv")
NAG_TRAIN = pd.read_csv("../data/processed/Nagpur/train.csv")
PIL_TRAIN = pd.read_csv("../data/processed/Pilani/train.csv")
```

```python
# Load Validation sets
DEL_VAL = pd.read_csv("../data/processed/Delhi/val.csv")
BLR_VAL = pd.read_csv("../data/processed/Bangalore/val.csv")
MUM_VAL = pd.read_csv("../data/processed/Mumbai/val.csv")
NAG_VAL = pd.read_csv("../data/processed/Nagpur/val.csv")
PIL_VAL = pd.read_csv("../data/processed/Pilani/val.csv")

# Load Test sets
DEL_TEST = pd.read_csv("../data/processed/Delhi/test.csv")
BLR_TEST = pd.read_csv("../data/processed/Bangalore/test.csv")
MUM_TEST = pd.read_csv("../data/processed/Mumbai/test.csv")
NAG_TEST = pd.read_csv("../data/processed/Nagpur/test.csv")
PIL_TEST = pd.read_csv("../data/processed/Pilani/test.csv")
```

```python
# Create unified datasets by concatenating city-specific DataFrames
Unified_Train_Set = pd.concat([DEL_TRAIN, BLR_TRAIN, MUM_TRAIN, NAG_TRAIN, PIL_TRAIN], ignore_index=True)
Unified_Val_Set = pd.concat([DEL_VAL, BLR_VAL, MUM_VAL, NAG_VAL, PIL_VAL], ignore_index=True)
Unified_Test_Set = pd.concat([DEL_TEST, BLR_TEST, MUM_TEST, NAG_TEST, PIL_TEST], ignore_index=True)

# Optional: verify the shapes
print(f"Train set shape: {Unified_Train_Set.shape}")
print(f"Val set shape: {Unified_Val_Set.shape}")
print(f"Test set shape: {Unified_Test_Set.shape}")
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error, accuracy_score, classification_report

# Create a working copy of the validation set
val_preds = Unified_Val_Set.copy()

# Sort by city and time (if there is a chronological order) to ensure lags calculate correctly
# If your rows aren't sorted chronologically, make sure to sort by date here first.
val_preds = val_preds.sort_values(by=['city_id', 'year', 'day']).reset_index(drop=True)

# -------------------------------------------------------------------------
# BACKBONE: FEATURE ENGINEERING (Lags & Rolling Metrics per City)
# -------------------------------------------------------------------------
print("Engineering lag and rolling features...")

# Temperature Lags & Rolling for Wind Direction Model
val_preds['temp_max_lag1'] = val_preds.groupby('city_id')['temp_max'].shift(1)
val_preds['temp_max_lag7'] = val_preds.groupby('city_id')['temp_max'].shift(7)
val_preds['temp_max_roll7'] = val_preds.groupby('city_id')['temp_max'].transform(lambda x: x.rolling(7, min_periods=1).mean())

val_preds['temp_min_lag1'] = val_preds.groupby('city_id')['temp_min'].shift(1)
val_preds['temp_min_lag7'] = val_preds.groupby('city_id')['temp_min'].shift(7)
val_preds['temp_min_roll7'] = val_preds.groupby('city_id')['temp_min'].transform(lambda x: x.rolling(7, min_periods=1).mean())

# Wind Speed Lags & Rolling for Wind Speed Model
val_preds['wind_speed_max_lag1'] = val_preds.groupby('city_id')['wind_speed_max'].shift(1)
val_preds['wind_speed_max_lag3'] = val_preds.groupby('city_id')['wind_speed_max'].shift(3)
val_preds['wind_speed_max_lag5'] = val_preds.groupby('city_id')['wind_speed_max'].shift(5)
val_preds['wind_speed_max_lag7'] = val_preds.groupby('city_id')['wind_speed_max'].shift(7)
val_preds['wind_speed_max_roll7'] = val_preds.groupby('city_id')['wind_speed_max'].transform(lambda x: x.rolling(7, min_periods=1).mean())
val_preds['wind_speed_max_std7'] = val_preds.groupby('city_id')['wind_speed_max'].transform(lambda x: x.rolling(7, min_periods=1).std().fillna(0))

# Wind Direction Sin/Cos Lags & Rolling for Wind Speed Model
# (Using the true historical wind direction components first)
val_preds['true_wind_dir_sin'] = np.sin(np.radians(val_preds['wind_dir_label'].astype(float)))
val_preds['true_wind_dir_cos'] = np.cos(np.radians(val_preds['wind_dir_label'].astype(float)))

for lag in [1, 3, 5, 7]:
    val_preds[f'wind_dir_sin_lag{lag}'] = val_preds.groupby('city_id')['true_wind_dir_sin'].shift(lag)
    val_preds[f'wind_dir_cos_lag{lag}'] = val_preds.groupby('city_id')['true_wind_dir_cos'].shift(lag)

val_preds['wind_dir_sin_roll7'] = val_preds.groupby('city_id')['true_wind_dir_sin'].transform(lambda x: x.rolling(7, min_periods=1).mean())
val_preds['wind_dir_sin_std7'] = val_preds.groupby('city_id')['true_wind_dir_sin'].transform(lambda x: x.rolling(7, min_periods=1).std().fillna(0))
val_preds['wind_dir_cos_roll7'] = val_preds.groupby('city_id')['true_wind_dir_cos'].transform(lambda x: x.rolling(7, min_periods=1).mean())
val_preds['wind_dir_cos_std7'] = val_preds.groupby('city_id')['true_wind_dir_cos'].transform(lambda x: x.rolling(7, min_periods=1).std().fillna(0))

# Drop any rows where shifting created NaNs (e.g., the first 7 days of data per city)
val_preds = val_preds.dropna().reset_index(drop=True)

# -------------------------------------------------------------------------
# 1. Predict Max & Min Temp 
# -------------------------------------------------------------------------
print("Predicting Temperatures...")
max_temp_features = ['year', 'month_sin', 'month_cos', 'day', 'dayofweek', 'city_id']
val_preds['temp_max_pred'] = temp_max_model.predict(val_preds[max_temp_features])

min_temp_features = ['year', 'month_sin', 'month_cos', 'day', 'dayofweek', 'city_id', 'temp_max_pred']
val_preds['temp_min_pred'] = temp_min_model.predict(val_preds[min_temp_features])

# -------------------------------------------------------------------------
# 2. Predict Wind Direction
# -------------------------------------------------------------------------
print("Predicting Wind Direction...")
wind_dir_features = [
    'city_id', 'weather_code', 'wind_speed_max', 'precip_sum', 'precip_hours',
    'year', 'day', 'dayofweek', 'month_sin', 'month_cos',
    'temp_max_lag1', 'temp_max_lag7', 'temp_max_roll7',
    'temp_min_lag1', 'temp_min_lag7', 'temp_min_roll7',
    'temp_max_pred', 'temp_min_pred'
]
val_preds['wind_dir_pred'] = wind_direction_model.predict(val_preds[wind_dir_features])

# -------------------------------------------------------------------------
# 3. Predict Wind Speed
# -------------------------------------------------------------------------
print("Predicting Wind Speed...")
val_preds['wind_dir_sin'] = np.sin(np.radians(val_preds['wind_dir_pred'].astype(float)))
val_preds['wind_dir_cos'] = np.cos(np.radians(val_preds['wind_dir_pred'].astype(float)))

val_preds['dayofyear'] = pd.to_datetime(val_preds['year'].astype(str) + '-' + val_preds['day'].astype(str), format='%Y-%d', errors='coerce').dt.dayofyear.fillna(1)
val_preds['dayofyear_sin'] = np.sin(2 * np.pi * val_preds['dayofyear'] / 365.25)
val_preds['dayofyear_cos'] = np.cos(2 * np.pi * val_preds['dayofyear'] / 365.25)

wind_speed_features = [
    'year', 'day', 'dayofweek', 'month_sin', 'month_cos', 'dayofyear_sin', 'dayofyear_cos',
    'city_id', 'temp_max', 'temp_min', 'wind_dir_sin', 'wind_dir_cos',
    'wind_speed_max_lag1', 'wind_speed_max_lag3', 'wind_speed_max_lag5', 'wind_speed_max_lag7', 'wind_speed_max_roll7', 'wind_speed_max_std7',
    'wind_dir_sin_lag1', 'wind_dir_sin_lag3', 'wind_dir_sin_lag5', 'wind_dir_sin_lag7', 'wind_dir_sin_roll7', 'wind_dir_sin_std7',
    'wind_dir_cos_lag1', 'wind_dir_cos_lag3', 'wind_dir_cos_lag5', 'wind_dir_cos_lag7', 'wind_dir_cos_roll7', 'wind_dir_cos_std7'
]
val_preds['wind_speed_pred'] = wind_speed_model.predict(val_preds[wind_speed_features])

print("Done! Predictions generated successfully.")
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, accuracy_score, classification_report

print("==================================================")
print("          WIND DIRECTION EVALUATION (Classifier)   ")
print("==================================================")
# Ensure data types match up for validation
y_true_dir = val_preds['wind_dir_label'].astype(int)
y_pred_dir = val_preds['wind_dir_pred'].astype(int)

print(f"Accuracy Score: {accuracy_score(y_true_dir, y_pred_dir):.4f}\n")
print("Classification Report:")
print(classification_report(y_true_dir, y_pred_dir))

print("\n==================================================")
print("          WIND SPEED EVALUATION (Regression)       ")
print("==================================================")
y_true_speed = val_preds['wind_speed_max']
y_pred_speed = val_preds['wind_speed_pred']

rmse_speed = np.sqrt(mean_squared_error(y_true_speed, y_pred_speed))
mae_speed = mean_absolute_error(y_true_speed, y_pred_speed)

print(f"Root Mean Squared Error (RMSE): {rmse_speed:.4f}")
print(f"Mean Absolute Error (MAE):      {mae_speed:.4f}")
```

```python
from sklearn.metrics import r2_score
import numpy as np

print("==================================================")
print("     CUSTOM WIND DIRECTION EVALUATION (±45°)      ")
print("==================================================")

# 1. Cast targets to integers
y_true_dir = val_preds['wind_dir_label'].astype(int).values
y_pred_dir = val_preds['wind_dir_pred'].astype(int).values

# 2. Calculate absolute class index distance
class_diff = np.abs(y_true_dir - y_pred_dir)

# 3. Account for modular wrap-around (e.g., true=7 and pred=0 is a diff of 1 class, not 7)
class_diff = np.minimum(class_diff, 8 - class_diff)

# 4. Success condition: difference is less than or equal to 1 class step (45 degrees)
within_45_deg = class_diff <= 1
dir_acc_45 = np.mean(within_45_deg)

print(f"Total Validation Samples:             {len(val_preds)}")
print(f"Predictions within ±45° (±1 class):   {np.sum(within_45_deg)}")
print(f"Directional Accuracy (±45° window):   {dir_acc_45:.4f} ({dir_acc_45 * 100:.2f}%)")


print("\n==================================================")
print("          WIND SPEED EVALUATION (R² Score)        ")
print("==================================================")

y_true_speed = val_preds['wind_speed_max']
y_pred_speed = val_preds['wind_speed_pred']

r2 = r2_score(y_true_speed, y_pred_speed)
print(f"R² (Coefficient of Determination):   {r2:.4f}")
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import r2_score

# -------------------------------------------------------------------------
# 1. RE-ENGINEER FEATURES CONCATENATING TRAIN + VAL FOR CORRECT LAGS
# -------------------------------------------------------------------------
print("Engineering true chronological lag and rolling features...")

# Combine both sets temporarily to prevent boundary NaNs in the Validation set
full_df = pd.concat([Unified_Train_Set, Unified_Val_Set], ignore_index=True)
full_df = full_df.sort_values(by=['city_id', 'year', 'day']).reset_index(drop=True)

# Temperature Lags & Rolling
full_df['temp_max_lag1'] = full_df.groupby('city_id')['temp_max'].shift(1)
full_df['temp_max_lag7'] = full_df.groupby('city_id')['temp_max'].shift(7)
full_df['temp_max_roll7'] = full_df.groupby('city_id')['temp_max'].transform(lambda x: x.rolling(7, min_periods=1).mean())

full_df['temp_min_lag1'] = full_df.groupby('city_id')['temp_min'].shift(1)
full_df['temp_min_lag7'] = full_df.groupby('city_id')['temp_min'].shift(7)
full_df['temp_min_roll7'] = full_df.groupby('city_id')['temp_min'].transform(lambda x: x.rolling(7, min_periods=1).mean())

# Wind Speed Lags & Rolling
full_df['wind_speed_max_lag1'] = full_df.groupby('city_id')['wind_speed_max'].shift(1)
full_df['wind_speed_max_lag3'] = full_df.groupby('city_id')['wind_speed_max'].shift(3)
full_df['wind_speed_max_lag5'] = full_df.groupby('city_id')['wind_speed_max'].shift(5)
full_df['wind_speed_max_lag7'] = full_df.groupby('city_id')['wind_speed_max'].shift(7)
full_df['wind_speed_max_roll7'] = full_df.groupby('city_id')['wind_speed_max'].transform(lambda x: x.rolling(7, min_periods=1).mean())
full_df['wind_speed_max_std7'] = full_df.groupby('city_id')['wind_speed_max'].transform(lambda x: x.rolling(7, min_periods=1).std().fillna(0))

# Wind Direction Sin/Cos Lags & Rolling (Using target labels converted to real degrees)
full_df['true_angles'] = full_df['wind_dir_label'].astype(float) * 45
full_df['true_wind_dir_sin'] = np.sin(np.radians(full_df['true_angles']))
full_df['true_wind_dir_cos'] = np.cos(np.radians(full_df['true_angles']))

for lag in [1, 3, 5, 7]:
    full_df[f'wind_dir_sin_lag{lag}'] = full_df.groupby('city_id')['true_wind_dir_sin'].shift(lag)
    full_df[f'wind_dir_cos_lag{lag}'] = full_df.groupby('city_id')['true_wind_dir_cos'].shift(lag)

full_df['wind_dir_sin_roll7'] = full_df.groupby('city_id')['true_wind_dir_sin'].transform(lambda x: x.rolling(7, min_periods=1).mean())
full_df['wind_dir_sin_std7'] = full_df.groupby('city_id')['true_wind_dir_sin'].transform(lambda x: x.rolling(7, min_periods=1).std().fillna(0))
full_df['wind_dir_cos_roll7'] = full_df.groupby('city_id')['true_wind_dir_cos'].transform(lambda x: x.rolling(7, min_periods=1).mean())
full_df['wind_dir_cos_std7'] = full_df.groupby('city_id')['true_wind_dir_cos'].transform(lambda x: x.rolling(7, min_periods=1).std().fillna(0))

# Chronological Day of Year components
full_df['dayofyear'] = pd.to_datetime(full_df['year'].astype(str) + '-' + full_df['day'].astype(str), format='%Y-%d', errors='coerce').dt.dayofyear.fillna(1)
full_df['dayofyear_sin'] = np.sin(2 * np.pi * full_df['dayofyear'] / 365.25)
full_df['dayofyear_cos'] = np.cos(2 * np.pi * full_df['dayofyear'] / 365.25)

# Slice back ONLY the validation set rows safely using original indicators
val_preds = full_df.iloc[len(Unified_Train_Set):].copy().reset_index(drop=True)

# -------------------------------------------------------------------------
# 2. RUN SEQUENTIAL PREDICTIONS (THE CHAINED PRODUCTION SCENARIO)
# -------------------------------------------------------------------------
print("Generating sequential predictions...")
max_temp_features = ['year', 'month_sin', 'month_cos', 'day', 'dayofweek', 'city_id']
val_preds['temp_max_pred'] = temp_max_model.predict(val_preds[max_temp_features])

min_temp_features = ['year', 'month_sin', 'month_cos', 'day', 'dayofweek', 'city_id', 'temp_max_pred']
val_preds['temp_min_pred'] = temp_min_model.predict(val_preds[min_temp_features])

wind_dir_features = [
    'city_id', 'weather_code', 'wind_speed_max', 'precip_sum', 'precip_hours',
    'year', 'day', 'dayofweek', 'month_sin', 'month_cos',
    'temp_max_lag1', 'temp_max_lag7', 'temp_max_roll7',
    'temp_min_lag1', 'temp_min_lag7', 'temp_min_roll7',
    'temp_max_pred', 'temp_min_pred'
]
val_preds['wind_dir_pred'] = wind_direction_model.predict(val_preds[wind_dir_features])

# Convert predicted direction class labels to real degrees for the Wind Speed model
pred_degrees = val_preds['wind_dir_pred'].astype(float) * 45
val_preds['wind_dir_sin'] = np.sin(np.radians(pred_degrees))
val_preds['wind_dir_cos'] = np.cos(np.radians(pred_degrees))

wind_speed_features = [
    'year', 'day', 'dayofweek', 'month_sin', 'month_cos', 'dayofyear_sin', 'dayofyear_cos',
    'city_id', 'temp_max', 'temp_min', 'wind_dir_sin', 'wind_dir_cos',
    'wind_speed_max_lag1', 'wind_speed_max_lag3', 'wind_speed_max_lag5', 'wind_speed_max_lag7', 'wind_speed_max_roll7', 'wind_speed_max_std7',
    'wind_dir_sin_lag1', 'wind_dir_sin_lag3', 'wind_dir_sin_lag5', 'wind_dir_sin_lag7', 'wind_dir_sin_roll7', 'wind_dir_sin_std7',
    'wind_dir_cos_lag1', 'wind_dir_cos_lag3', 'wind_dir_cos_lag5', 'wind_dir_cos_lag7', 'wind_dir_cos_roll7', 'wind_dir_cos_std7'
]
val_preds['wind_speed_pred'] = wind_speed_model.predict(val_preds[wind_speed_features])

# -------------------------------------------------------------------------
# 3. EVALUATION METRICS
# -------------------------------------------------------------------------
print("\n" + "="*50)
print("     FINAL EVALUATION RESULTS     ")
print("="*50)

# Wind Direction (±45 deg tolerance accuracy)
y_true_dir = val_preds['wind_dir_label'].astype(int).values
y_pred_dir = val_preds['wind_dir_pred'].astype(int).values
class_diff = np.abs(y_true_dir - y_pred_dir)
class_diff = np.minimum(class_diff, 8 - class_diff)
dir_acc_45 = np.mean(class_diff <= 1)

print(f"Wind Direction Accuracy (±45° Window): {dir_acc_45 * 100:.2f}%")

# Wind Speed Chained R² (using predicted values)
r2_chained = r2_score(val_preds['wind_speed_max'], val_preds['wind_speed_pred'])
print(f"Wind Speed R² (Chained Model Pipeline): {r2_chained:.4f}")

# -------------------------------------------------------------------------
# 4. ORACLE SANITY CHECK (EVALUATING MODEL CAPACITY WITH TRUE INPUTS)
# -------------------------------------------------------------------------
sanity_df = val_preds.copy()
true_deg_values = sanity_df['wind_dir_label'].astype(float) * 45
sanity_df['wind_dir_sin'] = np.sin(np.radians(true_deg_values))
sanity_df['wind_dir_cos'] = np.cos(np.radians(true_deg_values))

y_pred_speed_ideal = wind_speed_model.predict(sanity_df[wind_speed_features])

# FIX: Now cleanly references 'sanity_df' to resolve the variable sample length error
r2_ideal = r2_score(sanity_df['wind_speed_max'], y_pred_speed_ideal)
print(f"Wind Speed R² (Ideal Oracle Benchmark):  {r2_ideal:.4f}")
```

```python
import lightgbm as lgb
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score, classification_report
import pandas as pd
import numpy as np

# -------------------------------------------------------------------------
# 1. COMPREHENSIVE FEATURE ENGINEERING (Train + Val + Test Alignment)
# -------------------------------------------------------------------------
print("Engineering full suite of lag, rolling, and cyclical features...")

# Combine temporarily to prevent edge NaNs when shifting across dataset boundaries
full_df = pd.concat([Unified_Train_Set, Unified_Val_Set, Unified_Test_Set], ignore_index=True)
full_df = full_df.sort_values(by=['city_id', 'year', 'day']).reset_index(drop=True)

# A. Direction Angle Components
full_df['true_angles'] = full_df['wind_dir_label'].astype(float) * 45
full_df['wind_dir_sin'] = np.sin(np.radians(full_df['true_angles']))
full_df['wind_dir_cos'] = np.cos(np.radians(full_df['true_angles']))

# B. Wind Direction Lags (1, 3, 7)
for lag in [1, 3, 7]:
    full_df[f'wind_dir_sin_lag{lag}'] = full_df.groupby('city_id')['wind_dir_sin'].shift(lag)
    full_df[f'wind_dir_cos_lag{lag}'] = full_df.groupby('city_id')['wind_dir_cos'].shift(lag)

# C. Wind Speed Lags (1, 3, 7) & Rolling Average (7-day window)
for lag in [1, 3, 7]:
    full_df[f'wind_speed_max_lag{lag}'] = full_df.groupby('city_id')['wind_speed_max'].shift(lag)
full_df['wind_speed_max_roll7'] = full_df.groupby('city_id')['wind_speed_max'].transform(lambda x: x.rolling(7, min_periods=1).mean())

# D. Precip Sum Lags (1, 3, 7) & Rolling Average (7-day window)
for lag in [1, 3, 7]:
    full_df[f'precip_sum_lag{lag}'] = full_df.groupby('city_id')['precip_sum'].shift(lag)
full_df['precip_sum_roll7'] = full_df.groupby('city_id')['precip_sum'].transform(lambda x: x.rolling(7, min_periods=1).mean())

# E. Precip Hours Lags (1, 3, 7) & Rolling Average (7-day window)
for lag in [1, 3, 7]:
    full_df[f'precip_hours_lag{lag}'] = full_df.groupby('city_id')['precip_hours'].shift(lag)
full_df['precip_hours_roll7'] = full_df.groupby('city_id')['precip_hours'].transform(lambda x: x.rolling(7, min_periods=1).mean())

# F. Chronological Day of Year Cos/Sin components
full_df['dayofyear'] = pd.to_datetime(full_df['year'].astype(str) + '-' + full_df['day'].astype(str), format='%Y-%d', errors='coerce').dt.dayofyear.fillna(1)
full_df['dayofyear_sin'] = np.sin(2 * np.pi * full_df['dayofyear'] / 365.25)
full_df['dayofyear_cos'] = np.cos(2 * np.pi * full_df['dayofyear'] / 365.25)

# Slice datasets back safely and clean initial NaN rows created by shifts
train_idx = len(Unified_Train_Set)
val_idx = train_idx + len(Unified_Val_Set)

train_df = full_df.iloc[:train_idx].copy().dropna().reset_index(drop=True)
val_df = full_df.iloc[train_idx:val_idx].copy().dropna().reset_index(drop=True)

# -------------------------------------------------------------------------
# 2. DEFINE EXPANDED INPUT FEATURES LIST
# -------------------------------------------------------------------------
features_list = [
    'city_id', 'temp_max', 'temp_min', 
    'year', 'day', 'dayofweek', 'month_sin', 'month_cos', 'dayofyear_sin', 'dayofyear_cos',
    'wind_dir_sin_lag1', 'wind_dir_cos_lag1', 'wind_dir_sin_lag3', 'wind_dir_cos_lag3', 'wind_dir_sin_lag7', 'wind_dir_cos_lag7',
    'wind_speed_max_lag1', 'wind_speed_max_lag3', 'wind_speed_max_lag7', 'wind_speed_max_roll7',
    'precip_sum_lag1', 'precip_sum_lag3', 'precip_sum_lag7', 'precip_sum_roll7',
    'precip_hours_lag1', 'precip_hours_lag3', 'precip_hours_lag7', 'precip_hours_roll7'
]
target = 'wind_dir_label'

X_train = train_df[features_list]
y_train = train_df[target].astype(int)

X_val = val_df[features_list]
y_val = val_df[target].astype(int)

# -------------------------------------------------------------------------
# 3. INITIALIZE AND TRAIN THE MODEL
# -------------------------------------------------------------------------
print(f"Training updated model with {X_train.shape[1]} features...")
model = LGBMClassifier(
    objective='multiclass',
    n_estimators=3000,
    learning_rate=0.02,
    num_leaves=127,
    max_depth=10,
    min_child_samples=10,
    feature_fraction=0.9,
    bagging_fraction=0.9,
    bagging_freq=1,
    lambda_l1=0.0,
    lambda_l2=0.1,
    min_split_gain=0.0,
    max_bin=511,
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)

model.fit(
    X_train, y_train,
    eval_set=[(X_val, y_val)],
    callbacks=[
        lgb.early_stopping(100),
        lgb.log_evaluation(100)
    ]
)

# -------------------------------------------------------------------------
# 4. METRIC PERFORMANCE PRINTER
# -------------------------------------------------------------------------
print("\n" + "="*50)
print("       UPDATED MODEL ACCURACY RESULTS        ")
print("="*50)
y_true = y_val.values
y_pred = model.predict(X_val).astype(int)

exact_acc = accuracy_score(y_true, y_pred)
class_diff = np.abs(y_true - y_pred)
class_diff = np.minimum(class_diff, 8 - class_diff)
dir_acc_45 = np.mean(class_diff <= 1)

print(f"Exact Match Accuracy:                 {exact_acc * 100:.2f}%")
print(f"Directional Accuracy (±45° Window):   {dir_acc_45 * 100:.2f}%")

# -------------------------------------------------------------------------
# 5. NEW FEATURE IMPORTANCE (GAIN SORTED)
# -------------------------------------------------------------------------
print("\n==================================================")
print("     ALL FEATURES SORTED BY PREDICTIVE IMPACT     ")
print("==================================================")
importance_df = pd.DataFrame({
    'Feature': model.feature_name_,
    'Gain Importance': model.booster_.feature_importance(importance_type='gain'),
    'Split Count Importance': model.booster_.feature_importance(importance_type='split')
}).sort_values(by='Gain Importance', ascending=False).reset_index(drop=True)

pd.set_option('display.float_format', lambda x: '%.2f' % x)
print(importance_df.to_string(index=False))
```

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score, classification_report
import pandas as pd
import numpy as np

# -------------------------------------------------------------------------
# 1. COMPREHENSIVE FEATURE ENGINEERING (Train + Val + Test Alignment)
# -------------------------------------------------------------------------
print("Engineering full suite of lag, rolling, and cyclical features...")

full_df = pd.concat([Unified_Train_Set, Unified_Val_Set, Unified_Test_Set], ignore_index=True)
full_df = full_df.sort_values(by=['city_id', 'year', 'day']).reset_index(drop=True)

# A. Direction Angle Components
full_df['true_angles'] = full_df['wind_dir_label'].astype(float) * 45
full_df['wind_dir_sin'] = np.sin(np.radians(full_df['true_angles']))
full_df['wind_dir_cos'] = np.cos(np.radians(full_df['true_angles']))

# B. Wind Direction Lags (1, 3, 7)
for lag in [1, 3, 7]:
    full_df[f'wind_dir_sin_lag{lag}'] = full_df.groupby('city_id')['wind_dir_sin'].shift(lag)
    full_df[f'wind_dir_cos_lag{lag}'] = full_df.groupby('city_id')['wind_dir_cos'].shift(lag)

# C. Wind Speed Lags (1, 3, 7) & Rolling Average (7-day window)
for lag in [1, 3, 7]:
    full_df[f'wind_speed_max_lag{lag}'] = full_df.groupby('city_id')['wind_speed_max'].shift(lag)
full_df['wind_speed_max_roll7'] = full_df.groupby('city_id')['wind_speed_max'].transform(lambda x: x.rolling(7, min_periods=1).mean())

# D. Precip Sum Lags (1, 3, 7) & Rolling Average (7-day window)
for lag in [1, 3, 7]:
    full_df[f'precip_sum_lag{lag}'] = full_df.groupby('city_id')['precip_sum'].shift(lag)
full_df['precip_sum_roll7'] = full_df.groupby('city_id')['precip_sum'].transform(lambda x: x.rolling(7, min_periods=1).mean())

# E. Precip Hours Lags (1, 3, 7) & Rolling Average (7-day window)
for lag in [1, 3, 7]:
    full_df[f'precip_hours_lag{lag}'] = full_df.groupby('city_id')['precip_hours'].shift(lag)
full_df['precip_hours_roll7'] = full_df.groupby('city_id')['precip_hours'].transform(lambda x: x.rolling(7, min_periods=1).mean())

# F. Chronological Day of Year Cos/Sin components
full_df['dayofyear'] = pd.to_datetime(full_df['year'].astype(str) + '-' + full_df['day'].astype(str), format='%Y-%d', errors='coerce').dt.dayofyear.fillna(1)
full_df['dayofyear_sin'] = np.sin(2 * np.pi * full_df['dayofyear'] / 365.25)
full_df['dayofyear_cos'] = np.cos(2 * np.pi * full_df['dayofyear'] / 365.25)

# XGBoost requirement: Map categorical features strictly to numeric codes
if full_df['city_id'].dtype.name == 'category':
    full_df['city_id'] = full_df['city_id'].cat.codes
else:
    full_df['city_id'] = full_df['city_id'].astype(int)

# Slice datasets back safely and drop initial boundary NaNs created by shifts
train_idx = len(Unified_Train_Set)
val_idx = train_idx + len(Unified_Val_Set)

train_df = full_df.iloc[:train_idx].copy().dropna().reset_index(drop=True)
val_df = full_df.iloc[train_idx:val_idx].copy().dropna().reset_index(drop=True)

# -------------------------------------------------------------------------
# 2. DEFINE EXPANDED INPUT FEATURES LIST
# -------------------------------------------------------------------------
features_list = [
    'city_id', 'temp_max', 'temp_min', 
    'year', 'day', 'dayofweek', 'month_sin', 'month_cos', 'dayofyear_sin', 'dayofyear_cos',
    'wind_dir_sin_lag1', 'wind_dir_cos_lag1', 'wind_dir_sin_lag3', 'wind_dir_cos_lag3', 'wind_dir_sin_lag7', 'wind_dir_cos_lag7',
    'wind_speed_max_lag1', 'wind_speed_max_lag3', 'wind_speed_max_lag7', 'wind_speed_max_roll7',
    'precip_sum_lag1', 'precip_sum_lag3', 'precip_sum_lag7', 'precip_sum_roll7',
    'precip_hours_lag1', 'precip_hours_lag3', 'precip_hours_lag7', 'precip_hours_roll7'
]
target = 'wind_dir_label'

X_train = train_df[features_list].astype(float)
y_train = train_df[target].astype(int)

X_val = val_df[features_list].astype(float)
y_val = val_df[target].astype(int)

# -------------------------------------------------------------------------
# 3. INITIALIZE AND TRAIN THE XGBOOST MODEL
# -------------------------------------------------------------------------
print(f"Training XGBoost multiclass model with {X_train.shape[1]} features...")

# Calculate class weights manually since XGBoost doesn't have a simple 'balanced' string parameter
from sklearn.utils.class_weight import compute_sample_weight
sample_weights = compute_sample_weight(class_weight='balanced', y=y_train)

xgb_model = xgb.XGBClassifier(
    objective='multi:softprob',
    num_class=8,
    n_estimators=1000,
    learning_rate=0.05,
    max_depth=3,
    subsample=0.9,
    colsample_bytree=0.7,
    reg_alpha=1.0,
    reg_lambda=5.0,
    random_state=42,
    n_jobs=-1,
    early_stopping_rounds=100
)

xgb_model.fit(
    X_train, y_train,
    sample_weight=sample_weights,
    eval_set=[(X_val, y_val)],
    verbose=100
)

# -------------------------------------------------------------------------
# 4. METRIC PERFORMANCE PRINTER
# -------------------------------------------------------------------------
print("\n" + "="*50)
print("       XGBOOST MODEL ACCURACY RESULTS        ")
print("="*50)
y_true = y_val.values
y_pred = xgb_model.predict(X_val).astype(int)

exact_acc = accuracy_score(y_true, y_pred)
class_diff = np.abs(y_true - y_pred)
class_diff = np.minimum(class_diff, 8 - class_diff)
dir_acc_45 = np.mean(class_diff <= 1)

print(f"Exact Match Accuracy:                 {exact_acc * 100:.2f}%")
print(f"Directional Accuracy (±45° Window):   {dir_acc_45 * 100:.2f}%")

# -------------------------------------------------------------------------
# 5. NEW XGBOOST FEATURE IMPORTANCE (WEIGHT/GAIN)
# -------------------------------------------------------------------------
print("\n==================================================")
print("     ALL FEATURES SORTED BY PREDICTIVE IMPACT     ")
print("==================================================")
# XGBoost provides gains inside a dictionary mapping
importance_dict = xgb_model.get_booster().get_score(importance_type='gain')

importance_df = pd.DataFrame({
    'Feature': features_list,
    'Gain Importance': [importance_dict.get(f, 0.0) for f in features_list]
}).sort_values(by='Gain Importance', ascending=False).reset_index(drop=True)

print(importance_df.to_string(index=False))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import pandas as pd
import numpy as np

# -------------------------------------------------------------------------
# 1. CHRONOLOGICAL FEATURE ENGINEERING FOR WIND SPEED
# -------------------------------------------------------------------------
print("Engineering full chronological features for Wind Speed evaluation...")

# Combine all sets temporarily to ensure lags and rolling windows are perfectly continuous
full_df = pd.concat([Unified_Train_Set, Unified_Val_Set, Unified_Test_Set], ignore_index=True)
full_df = full_df.sort_values(by=['city_id', 'year', 'day']).reset_index(drop=True)

# A. Convert target labels (0-7) to real degrees and extract components
full_df['true_angles'] = full_df['wind_dir_label'].astype(float) * 45
full_df['wind_dir_sin'] = np.sin(np.radians(full_df['true_angles']))
full_df['wind_dir_cos'] = np.cos(np.radians(full_df['true_angles']))

# B. Generate Wind Direction Lags (1, 3, 5, 7)
for lag in [1, 3, 5, 7]:
    full_df[f'wind_dir_sin_lag{lag}'] = full_df.groupby('city_id')['wind_dir_sin'].shift(lag)
    full_df[f'wind_dir_cos_lag{lag}'] = full_df.groupby('city_id')['wind_dir_cos'].shift(lag)

# C. Generate Wind Direction Rolling Averages & Standard Deviations
full_df['wind_dir_sin_roll7'] = full_df.groupby('city_id')['wind_dir_sin'].transform(lambda x: x.rolling(7, min_periods=1).mean())
full_df['wind_dir_sin_std7'] = full_df.groupby('city_id')['wind_dir_sin'].transform(lambda x: x.rolling(7, min_periods=1).std().fillna(0))
full_df['wind_dir_cos_roll7'] = full_df.groupby('city_id')['wind_dir_cos'].transform(lambda x: x.rolling(7, min_periods=1).mean())
full_df['wind_dir_cos_std7'] = full_df.groupby('city_id')['wind_dir_cos'].transform(lambda x: x.rolling(7, min_periods=1).std().fillna(0))

# D. Generate Wind Speed Lags (1, 3, 5, 7) & Statistics
for lag in [1, 3, 5, 7]:
    full_df[f'wind_speed_max_lag{lag}'] = full_df.groupby('city_id')['wind_speed_max'].shift(lag)
full_df['wind_speed_max_roll7'] = full_df.groupby('city_id')['wind_speed_max'].transform(lambda x: x.rolling(7, min_periods=1).mean())
full_df['wind_speed_max_std7'] = full_df.groupby('city_id')['wind_speed_max'].transform(lambda x: x.rolling(7, min_periods=1).std().fillna(0))

# E. Generate Day of Year cyclical features
full_df['dayofyear'] = pd.to_datetime(full_df['year'].astype(str) + '-' + full_df['day'].astype(str), format='%Y-%d', errors='coerce').dt.dayofyear.fillna(1)
full_df['dayofyear_sin'] = np.sin(2 * np.pi * full_df['dayofyear'] / 365.25)
full_df['dayofyear_cos'] = np.cos(2 * np.pi * full_df['dayofyear'] / 365.25)

# -------------------------------------------------------------------------
# 2. SEPARATE INTO SPECIFIC EVALUATION SPLITS
# -------------------------------------------------------------------------
train_len = len(Unified_Train_Set)
val_len = len(Unified_Val_Set)

# Extract fresh copies of validation and test segments
eval_val_df = full_df.iloc[train_len : train_len + val_len].copy().dropna().reset_index(drop=True)
eval_test_df = full_df.iloc[train_len + val_len :].copy().dropna().reset_index(drop=True)

# Define features expected by your wind speed model
wind_speed_features = [
    'year', 'day', 'dayofweek', 'month_sin', 'month_cos', 'dayofyear_sin', 'dayofyear_cos',
    'city_id', 'temp_max', 'temp_min', 'wind_dir_sin', 'wind_dir_cos',
    'wind_speed_max_lag1', 'wind_speed_max_lag3', 'wind_speed_max_lag5', 'wind_speed_max_lag7', 'wind_speed_max_roll7', 'wind_speed_max_std7',
    'wind_dir_sin_lag1', 'wind_dir_sin_lag3', 'wind_dir_sin_lag5', 'wind_dir_sin_lag7', 'wind_dir_sin_roll7', 'wind_dir_sin_std7',
    'wind_dir_cos_lag1', 'wind_dir_cos_lag3', 'wind_dir_cos_lag5', 'wind_dir_cos_lag7', 'wind_dir_cos_roll7', 'wind_dir_cos_std7'
]
target = 'wind_speed_max'

# -------------------------------------------------------------------------
# 3. RUN EVALUATION: VALIDATION SET
# -------------------------------------------------------------------------
print("Evaluating on Validation Set...")
y_true_val = eval_val_df[target].values
y_pred_val = wind_speed_model.predict(eval_val_df[wind_speed_features])

rmse_val = np.sqrt(mean_squared_error(y_true_val, y_pred_val))
mae_val = mean_absolute_error(y_true_val, y_pred_val)
r2_val = r2_score(y_true_val, y_pred_val)

# -------------------------------------------------------------------------
# 4. RUN EVALUATION: UNSEEN TEST SET
# -------------------------------------------------------------------------
print("Evaluating on Unseen Test Set...")
y_true_test = eval_test_df[target].values
y_pred_test = wind_speed_model.predict(eval_test_df[wind_speed_features])

rmse_test = np.sqrt(mean_squared_error(y_true_test, y_pred_test))
mae_test = mean_absolute_error(y_true_test, y_pred_test)
r2_test = r2_score(y_true_test, y_pred_test)

# -------------------------------------------------------------------------
# 5. RENDER FINAL PERFORMANCE REPORT
# -------------------------------------------------------------------------
print("\n" + "="*50)
# Final metrics output
print("          WIND SPEED MODEL BENCHMARK REPORT       ")
print("="*50)

performance_summary = pd.DataFrame({
    'Metric': ['R² Score', 'Root Mean Squared Error (RMSE)', 'Mean Absolute Error (MAE)'],
    'Validation Set': [r2_val, rmse_val, mae_val],
    'Unseen Test Set': [r2_test, rmse_test, mae_test]
})

print(performance_summary.to_string(index=False))
print("="*50)
```

```python
import lightgbm as lgb
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_absolute_error, r2_score
import pandas as pd
import numpy as np

# -------------------------------------------------------------------------
# 1. CHRONOLOGICAL FEATURE ENGINEERING & VECTOR AVERAGING
# -------------------------------------------------------------------------
print("Engineering topography anchors and directional vector means...")

# Temporarily combine to cleanly compute chronological windows across splits
full_df = pd.concat([Unified_Train_Set, Unified_Val_Set, Unified_Test_Set], ignore_index=True)
full_df = full_df.sort_values(by=['city_id', 'year', 'day']).reset_index(drop=True)

# Requirement A: Topography Anchor (Average wind speed per city)
full_df['city_mean_wind'] = full_df.groupby('city_id')['wind_speed_max'].transform('mean')

# Convert predicted/true direction labels to real radians for exact circular math
full_df['true_angles_rad'] = np.radians(full_df['wind_dir_label'].astype(float) * 45)
full_df['wind_dir_sin'] = np.sin(full_df['true_angles_rad'])
full_df['wind_dir_cos'] = np.cos(full_df['true_angles_rad'])

# Requirement B: Replace 18 lags with a 7-day Moving Direction Vector Mean
# We sum the sine and cosine components over the last 7 days to get the resultant vector
sum_sin_7d = full_df.groupby('city_id')['wind_dir_sin'].transform(lambda x: x.rolling(7, min_periods=1).sum())
sum_cos_7d = full_df.groupby('city_id')['wind_dir_cos'].transform(lambda x: x.rolling(7, min_periods=1).sum())

# Compute the clean mean directional angles using arctan2
full_df['vector_mean_dir_sin'] = sum_sin_7d / 7.0
full_df['vector_mean_dir_cos'] = sum_cos_7d / 7.0

# Requirement C: Keep only Lag 1 and Roll 7 for Wind Speed
full_df['wind_speed_max_lag1'] = full_df.groupby('city_id')['wind_speed_max'].shift(1)
full_df['wind_speed_max_roll7'] = full_df.groupby('city_id')['wind_speed_max'].transform(lambda x: x.rolling(7, min_periods=1).mean())

# Cyclical day of year components
full_df['dayofyear'] = pd.to_datetime(full_df['year'].astype(str) + '-' + full_df['day'].astype(str), format='%Y-%d', errors='coerce').dt.dayofyear.fillna(1)
full_df['dayofyear_sin'] = np.sin(2 * np.pi * full_df['dayofyear'] / 365.25)
full_df['dayofyear_cos'] = np.cos(2 * np.pi * full_df['dayofyear'] / 365.25)

# Slice datasets back safely and clean boundary NaNs from lag1
train_len = len(Unified_Train_Set)
val_len = len(Unified_Val_Set)

train_df = full_df.iloc[:train_len].copy().dropna().reset_index(drop=True)
val_df = full_df.iloc[train_len : train_len + val_len].copy().dropna().reset_index(drop=True)

# -------------------------------------------------------------------------
# 2. DEFINE FIXED PARAMETERIZED FEATURE SELECTION
# -------------------------------------------------------------------------
features_list = [
    'year', 'day', 'dayofweek', 'month_sin', 'month_cos', 'dayofyear_sin', 'dayofyear_cos',
    'city_id', 'temp_max', 'temp_min', 'wind_dir_sin', 'wind_dir_cos',
    'city_mean_wind',         # Spatial topography anchor
    'vector_mean_dir_sin',    # Compressed vector components
    'vector_mean_dir_cos',
    'wind_speed_max_lag1',    # Only remaining speed history items
    'wind_speed_max_roll7'
]
target = 'wind_speed_max'

X_train = train_df[features_list]
y_train = train_df[target].astype(float)

X_val = val_df[features_list]
y_val = val_df[target].astype(float)

# -------------------------------------------------------------------------
# 3. INITIALIZE AND TRAIN NEW TWEEDIE MODEL
# -------------------------------------------------------------------------
print("Training new Tweedie Wind Speed model...")

tweedie_speed_model = LGBMRegressor(
    objective='tweedie',
    tweedie_variance_power=1.3, # Centered right between 1.2 and 1.4
    n_estimators=2000,
    learning_rate=0.05,
    max_depth=5,                # Limited max tree depth spectrum
    num_leaves=31,              # Adjusted to complement a max_depth of 5 (2^5)
    colsample_bytree=0.6,       # Subsample feature fraction per tree
    random_state=42,
    n_jobs=-1
)

tweedie_speed_model.fit(
    X_train, y_train,
    eval_set=[(X_val, y_val)],
    callbacks=[lgb.early_stopping(50, verbose=False)]
)

# -------------------------------------------------------------------------
# 4. COMPUTE AND CHECK NEW MAE AND R² METRICS
# -------------------------------------------------------------------------
print("\n" + "="*50)
print("       NEW TWEEDIE SPEED PIPELINE PERFORMANCE     ")
print("="*50)

y_pred_val = tweedie_speed_model.predict(X_val)

new_mae = mean_absolute_error(y_val, y_pred_val)
new_r2 = r2_score(y_val, y_pred_val)

print(f"New Validation Mean Absolute Error (MAE): {new_mae:.4f}")
print(f"New Validation R² Score:                  {new_r2:.4f}")
print("="*50)
```

```python
import lightgbm as lgb
from lightgbm import LGBMRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, r2_score
import pandas as pd
import numpy as np

# -------------------------------------------------------------------------
# 1. DATA PREPARATION (Extracting All 3 Splits Chronologically)
# -------------------------------------------------------------------------
print("Preparing chronological features for the ensemble...")

# Combine temporarily to compute seamless moving windows across dataset splits
full_df = pd.concat([Unified_Train_Set, Unified_Val_Set, Unified_Test_Set], ignore_index=True)
full_df = full_df.sort_values(by=['city_id', 'year', 'day']).reset_index(drop=True)

# Topography Anchor
full_df['city_mean_wind'] = full_df.groupby('city_id')['wind_speed_max'].transform('mean')

# Moving Direction Vector Mean
full_df['true_angles_rad'] = np.radians(full_df['wind_dir_label'].astype(float) * 45)
full_df['wind_dir_sin'] = np.sin(full_df['true_angles_rad'])
full_df['wind_dir_cos'] = np.cos(full_df['true_angles_rad'])
sum_sin_7d = full_df.groupby('city_id')['wind_dir_sin'].transform(lambda x: x.rolling(7, min_periods=1).sum())
sum_cos_7d = full_df.groupby('city_id')['wind_dir_cos'].transform(lambda x: x.rolling(7, min_periods=1).sum())
full_df['vector_mean_dir_sin'] = sum_sin_7d / 7.0
full_df['vector_mean_dir_cos'] = sum_cos_7d / 7.0

# Remaining Speed History
full_df['wind_speed_max_lag1'] = full_df.groupby('city_id')['wind_speed_max'].shift(1)
full_df['wind_speed_max_roll7'] = full_df.groupby('city_id')['wind_speed_max'].transform(lambda x: x.rolling(7, min_periods=1).mean())

# Day of Year Cyclical components
full_df['dayofyear'] = pd.to_datetime(full_df['year'].astype(str) + '-' + full_df['day'].astype(str), format='%Y-%d', errors='coerce').dt.dayofyear.fillna(1)
full_df['dayofyear_sin'] = np.sin(2 * np.pi * full_df['dayofyear'] / 365.25)
full_df['dayofyear_cos'] = np.cos(2 * np.pi * full_df['dayofyear'] / 365.25)

# Strictly handle categorization codes for Random Forest compatibility
if full_df['city_id'].dtype.name == 'category':
    full_df['city_id'] = full_df['city_id'].cat.codes

# Split back cleanly and drop lag boundary artifacts
train_len = len(Unified_Train_Set)
val_len = len(Unified_Val_Set)

train_df = full_df.iloc[:train_len].copy().dropna().reset_index(drop=True)
val_df = full_df.iloc[train_len : train_len + val_len].copy().dropna().reset_index(drop=True)
test_df = full_df.iloc[train_len + val_len:].copy().dropna().reset_index(drop=True)

features_list = [
    'year', 'day', 'dayofweek', 'month_sin', 'month_cos', 'dayofyear_sin', 'dayofyear_cos',
    'city_id', 'temp_max', 'temp_min', 'wind_dir_sin', 'wind_dir_cos',
    'city_mean_wind', 'vector_mean_dir_sin', 'vector_mean_dir_cos',
    'wind_speed_max_lag1', 'wind_speed_max_roll7'
]
target = 'wind_speed_max'

X_train, y_train = train_df[features_list], train_df[target].astype(float)
X_val, y_val = val_df[features_list], val_df[target].astype(float)
X_test, y_test = test_df[features_list], test_df[target].astype(float)

# -------------------------------------------------------------------------
# 2. INITIALIZE AND TRAIN THE 3 BASE MODELS
# -------------------------------------------------------------------------
print("\nTraining Base Models...")

# Model A: Tweedie Regressor (Mean Trend)
model_A = LGBMRegressor(objective='tweedie', tweedie_variance_power=1.3, n_estimators=1000, 
                        learning_rate=0.05, max_depth=5, num_leaves=31, colsample_bytree=0.6, random_state=42, n_jobs=-1)
model_A.fit(X_train, y_train, eval_set=[(X_val, y_val)], callbacks=[lgb.early_stopping(50, verbose=False)])

# Model B: Quantile Regressor (Upper Bound / Gust Potential at 0.90 Quantile)
model_B = LGBMRegressor(objective='quantile', alpha=0.90, n_estimators=1000, 
                        learning_rate=0.05, max_depth=5, num_leaves=31, colsample_bytree=0.6, random_state=42, n_jobs=-1)
model_B.fit(X_train, y_train, eval_set=[(X_val, y_val)], callbacks=[lgb.early_stopping(50, verbose=False)])

# Model C: Random Forest Regressor (Non-linear Topographical Noise)
model_C = RandomForestRegressor(n_estimators=100, max_depth=8, min_samples_leaf=5, random_state=42, n_jobs=-1)
model_C.fit(X_train, y_train)

# -------------------------------------------------------------------------
# 3. BUILD META-FEATURES & TRAIN RIDGE META-LEARNER
# -------------------------------------------------------------------------
print("Training Ridge Meta-Learner...")

# Generate predictions on Train set to serve as training inputs for the Meta-Learner
meta_X_train = pd.DataFrame({
    'pred_A': model_A.predict(X_train),
    'pred_B': model_B.predict(X_train),
    'pred_C': model_C.predict(X_train)
})

# Initialize and train the final Stacking Meta-Learner
meta_learner = Ridge(alpha=1.0, positive=True) # positive=True preserves the directional correlation
meta_learner.fit(meta_X_train, y_train)

# -------------------------------------------------------------------------
# 4. GENERATE ENSEMBLE PREDICTIONS ACROSS ALL SPLITS
# -------------------------------------------------------------------------
# Train Ensemble Preds
y_pred_train = meta_learner.predict(meta_X_train)

# Val Ensemble Preds
meta_X_val = pd.DataFrame({
    'pred_A': model_A.predict(X_val),
    'pred_B': model_B.predict(X_val),
    'pred_C': model_C.predict(X_val)
})
y_pred_val = meta_learner.predict(meta_X_val)

# Test Ensemble Preds
meta_X_test = pd.DataFrame({
    'pred_A': model_A.predict(X_test),
    'pred_B': model_B.predict(X_test),
    'pred_C': model_C.predict(X_test)
})
y_pred_test = meta_learner.predict(meta_X_test)

# -------------------------------------------------------------------------
# 5. EVALUATE AND COMPARE RESULTS
# -------------------------------------------------------------------------
print("\n" + "="*60)
print("     STACKED ENSEMBLE SYSTEM PERFORMANCE REPORT     ")
print("="*60)

comparison_metrics = pd.DataFrame({
    'Dataset Split': ['Training Set', 'Validation Set', 'Unseen Test Set'],
    'MAE': [
        mean_absolute_error(y_train, y_pred_train),
        mean_absolute_error(y_val, y_pred_val),
        mean_absolute_error(y_test, y_pred_test)
    ],
    'R² Score': [
        r2_score(y_train, y_pred_train),
        r2_score(y_val, y_pred_val),
        r2_score(y_test, y_pred_test)
    ]
})

print(comparison_metrics.to_string(index=False))
print("-" * 60)
print("Meta-Learner Feature Weights (Blending Ratios):")
for feat, coef in zip(meta_X_train.columns, meta_learner.coef_):
    print(f" -> {feat.replace('pred_', 'Model ')}: {coef:.4f}")
print("="*60)
```

```python
import lightgbm as lgb
from lightgbm import LGBMRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score
import pandas as pd
import numpy as np

# -------------------------------------------------------------------------
# 1. DATA PREPARATION (Extracting All 3 Splits Chronologically)
# -------------------------------------------------------------------------
print("Preparing chronological features for Residual Boosting...")

full_df = pd.concat([Unified_Train_Set, Unified_Val_Set, Unified_Test_Set], ignore_index=True)
full_df = full_df.sort_values(by=['city_id', 'year', 'day']).reset_index(drop=True)

# Topography Anchor
full_df['city_mean_wind'] = full_df.groupby('city_id')['wind_speed_max'].transform('mean')

# Moving Direction Vector Mean
full_df['true_angles_rad'] = np.radians(full_df['wind_dir_label'].astype(float) * 45)
full_df['wind_dir_sin'] = np.sin(full_df['true_angles_rad'])
full_df['wind_dir_cos'] = np.cos(full_df['true_angles_rad'])
sum_sin_7d = full_df.groupby('city_id')['wind_dir_sin'].transform(lambda x: x.rolling(7, min_periods=1).sum())
sum_cos_7d = full_df.groupby('city_id')['wind_dir_cos'].transform(lambda x: x.rolling(7, min_periods=1).sum())
full_df['vector_mean_dir_sin'] = sum_sin_7d / 7.0
full_df['vector_mean_dir_cos'] = sum_cos_7d / 7.0

# Remaining Speed History
full_df['wind_speed_max_lag1'] = full_df.groupby('city_id')['wind_speed_max'].shift(1)
full_df['wind_speed_max_roll7'] = full_df.groupby('city_id')['wind_speed_max'].transform(lambda x: x.rolling(7, min_periods=1).mean())

# Day of Year Cyclical components
full_df['dayofyear'] = pd.to_datetime(full_df['year'].astype(str) + '-' + full_df['day'].astype(str), format='%Y-%d', errors='coerce').dt.dayofyear.fillna(1)
full_df['dayofyear_sin'] = np.sin(2 * np.pi * full_df['dayofyear'] / 365.25)
full_df['dayofyear_cos'] = np.cos(2 * np.pi * full_df['dayofyear'] / 365.25)

if full_df['city_id'].dtype.name == 'category':
    full_df['city_id'] = full_df['city_id'].cat.codes

train_len = len(Unified_Train_Set)
val_len = len(Unified_Val_Set)

train_df = full_df.iloc[:train_len].copy().dropna().reset_index(drop=True)
val_df = full_df.iloc[train_len : train_len + val_len].copy().dropna().reset_index(drop=True)
test_df = full_df.iloc[train_len + val_len:].copy().dropna().reset_index(drop=True)

features_list = [
    'year', 'day', 'dayofweek', 'month_sin', 'month_cos', 'dayofyear_sin', 'dayofyear_cos',
    'city_id', 'temp_max', 'temp_min', 'wind_dir_sin', 'wind_dir_cos',
    'city_mean_wind', 'vector_mean_dir_sin', 'vector_mean_dir_cos',
    'wind_speed_max_lag1', 'wind_speed_max_roll7'
]
target = 'wind_speed_max'

X_train, y_train = train_df[features_list], train_df[target].astype(float).values
X_val, y_val = val_df[features_list], val_df[target].astype(float).values
X_test, y_test = test_df[features_list], test_df[target].astype(float).values

# -------------------------------------------------------------------------
# 2. SEQUENTIAL RESIDUAL BOOSTING TRAINING
# -------------------------------------------------------------------------
print("\n--- Training Step 1: Base Tweedie Trend (Model A) ---")
model_A = LGBMRegressor(objective='tweedie', tweedie_variance_power=1.3, n_estimators=1000, 
                        learning_rate=0.05, max_depth=5, num_leaves=31, colsample_bytree=0.6, random_state=42, n_jobs=-1)
model_A.fit(X_train, y_train, eval_set=[(X_val, y_val)], callbacks=[lgb.early_stopping(50, verbose=False)])

# Compute Residuals after Model A
pred_A_train = model_A.predict(X_train)
pred_A_val = model_A.predict(X_val)
res_1_train = y_train - pred_A_train
res_1_val = y_val - pred_A_val

print("--- Training Step 2: Quantile Gust Adjustment on Residuals (Model B) ---")
model_B = LGBMRegressor(objective='quantile', alpha=0.90, n_estimators=1000, 
                        learning_rate=0.05, max_depth=5, num_leaves=31, colsample_bytree=0.6, random_state=42, n_jobs=-1)
model_B.fit(X_train, res_1_train, eval_set=[(X_val, res_1_val)], callbacks=[lgb.early_stopping(50, verbose=False)])

# Compute Residuals after Model B
pred_B_train = model_B.predict(X_train)
pred_B_val = model_B.predict(X_val)
res_2_train = res_1_train - pred_B_train

print("--- Training Step 3: Random Forest Structural Cleanup on Remaining Noise (Model C) ---")
model_C = RandomForestRegressor(n_estimators=100, max_depth=8, min_samples_leaf=5, random_state=42, n_jobs=-1)
model_C.fit(X_train, res_2_train)

# -------------------------------------------------------------------------
# 3. GENERATE FINAL ADDITIVE PREDICTIONS ACROSS ALL SPLITS
# -------------------------------------------------------------------------
# Train Set
y_pred_train = model_A.predict(X_train) + model_B.predict(X_train) + model_C.predict(X_train)

# Validation Set
y_pred_val = model_A.predict(X_val) + model_B.predict(X_val) + model_C.predict(X_val)

# Test Set
y_pred_test = model_A.predict(X_test) + model_B.predict(X_test) + model_C.predict(X_test)

# -------------------------------------------------------------------------
# 4. EVALUATE PERFORMANCE
# -------------------------------------------------------------------------
print("\n" + "="*60)
print("     RESIDUAL BOOSTING ENSEMBLE PERFORMANCE REPORT     ")
print("="*60)

comparison_metrics = pd.DataFrame({
    'Dataset Split': ['Training Set', 'Validation Set', 'Unseen Test Set'],
    'MAE': [
        mean_absolute_error(y_train, y_pred_train),
        mean_absolute_error(y_val, y_pred_val),
        mean_absolute_error(y_test, y_pred_test)
    ],
    'R² Score': [
        r2_score(y_train, y_pred_train),
        r2_score(y_val, y_pred_val),
        r2_score(y_test, y_pred_test)
    ]
})

print(comparison_metrics.to_string(index=False))
print("="*60)
```

```python
import lightgbm as lgb
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_absolute_error, r2_score
import pandas as pd
import numpy as np

# -------------------------------------------------------------------------
# 1. CHRONOLOGICAL FEATURE ENGINEERING & CITY NORMALIZATION
# -------------------------------------------------------------------------
print("Engineering features and applying city-wise wind speed normalization...")

# Combine temporarily to compute seamless chronological features across splits
full_df = pd.concat([Unified_Train_Set, Unified_Val_Set, Unified_Test_Set], ignore_index=True)
full_df = full_df.sort_values(by=['city_id', 'year', 'day']).reset_index(drop=True)

# A. Calculate Topology Anchor
full_df['city_mean_wind'] = full_df.groupby('city_id')['wind_speed_max'].transform('mean')

# B. Target Normalization: Scale wind speed relative to each city's unique topography
# We add a tiny epsilon (1e-5) to prevent potential division by zero
full_df['normalized_wind_speed'] = full_df['wind_speed_max'] / (full_df['city_mean_wind'] + 1e-5)

# C. 7-day Moving Direction Vector Mean
full_df['true_angles_rad'] = np.radians(full_df['wind_dir_label'].astype(float) * 45)
full_df['wind_dir_sin'] = np.sin(full_df['true_angles_rad'])
full_df['wind_dir_cos'] = np.cos(full_df['true_angles_rad'])

sum_sin_7d = full_df.groupby('city_id')['wind_dir_sin'].transform(lambda x: x.rolling(7, min_periods=1).sum())
sum_cos_7d = full_df.groupby('city_id')['wind_dir_cos'].transform(lambda x: x.rolling(7, min_periods=1).sum())
full_df['vector_mean_dir_sin'] = sum_sin_7d / 7.0
full_df['vector_mean_dir_cos'] = sum_cos_7d / 7.0

# D. Compressed Speed History (Lag 1 and Roll 7 only)
full_df['wind_speed_max_lag1'] = full_df.groupby('city_id')['wind_speed_max'].shift(1)
full_df['wind_speed_max_roll7'] = full_df.groupby('city_id')['wind_speed_max'].transform(lambda x: x.rolling(7, min_periods=1).mean())

# E. Cyclical Day of Year Components
full_df['dayofyear'] = pd.to_datetime(full_df['year'].astype(str) + '-' + full_df['day'].astype(str), format='%Y-%d', errors='coerce').dt.dayofyear.fillna(1)
full_df['dayofyear_sin'] = np.sin(2 * np.pi * full_df['dayofyear'] / 365.25)
full_df['dayofyear_cos'] = np.cos(2 * np.pi * full_df['dayofyear'] / 365.25)

if full_df['city_id'].dtype.name == 'category':
    full_df['city_id'] = full_df['city_id'].cat.codes

# Slice back into clean splits and remove edge artifact rows
train_len = len(Unified_Train_Set)
val_len = len(Unified_Val_Set)

train_df = full_df.iloc[:train_len].copy().dropna().reset_index(drop=True)
val_df = full_df.iloc[train_len : train_len + val_len].copy().dropna().reset_index(drop=True)
test_df = full_df.iloc[train_len + val_len:].copy().dropna().reset_index(drop=True)

# -------------------------------------------------------------------------
# 2. FEATURE SELECTION & TARGET SETUP
# -------------------------------------------------------------------------
features_list = [
    'year', 'day', 'dayofweek', 'month_sin', 'month_cos', 'dayofyear_sin', 'dayofyear_cos',
    'city_id', 'temp_max', 'temp_min', 'wind_dir_sin', 'wind_dir_cos',
    'city_mean_wind', 'vector_mean_dir_sin', 'vector_mean_dir_cos',
    'wind_speed_max_lag1', 'wind_speed_max_roll7'
]

X_train = train_df[features_list]
X_val = val_df[features_list]
X_test = test_df[features_list]

# Training target is the normalized scale variant
y_train_norm = train_df['normalized_wind_speed'].values
y_val_norm = val_df['normalized_wind_speed'].values

# Keep original ground truth unscaled values for final true evaluation
y_train_true = train_df['wind_speed_max'].values
y_val_true = val_df['wind_speed_max'].values
y_test_true = test_df['wind_speed_max'].values

# -------------------------------------------------------------------------
# 3. TRAINING SINGLE TWEEDIE MODEL
# -------------------------------------------------------------------------
print("Training simplified Pure Tweedie Regressor...")
tweedie_model = LGBMRegressor(
    objective='tweedie',
    tweedie_variance_power=1.3,
    n_estimators=1500,
    learning_rate=0.05,
    max_depth=5,
    num_leaves=31,
    colsample_bytree=0.6,
    random_state=42,
    n_jobs=-1
)

tweedie_model.fit(
    X_train, y_train_norm,
    eval_set=[(X_val, y_val_norm)],
    callbacks=[lgb.early_stopping(50, verbose=False)]
)

# -------------------------------------------------------------------------
# 4. INVERSE TRANSFORMATION & PERFORMANCE METRICS
# -------------------------------------------------------------------------
# Helper function to predict normalized factor and map back to real wind speed units
def get_denormalized_predictions(model_instance, features, city_means):
    pred_norm = model_instance.predict(features)
    return pred_norm * city_means

# Generate unscaled physical predictions
y_pred_train = get_denormalized_predictions(tweedie_model, X_train, train_df['city_mean_wind'].values)
y_pred_val = get_denormalized_predictions(tweedie_model, X_val, val_df['city_mean_wind'].values)
y_pred_test = get_denormalized_predictions(tweedie_model, X_test, test_df['city_mean_wind'].values)

# -------------------------------------------------------------------------
# 5. PRINT BENCHMARK COMPARISON
# -------------------------------------------------------------------------
print("\n" + "="*60)
print("     PURE TWEEDIE WITH CITY NORMALIZATION METRICS       ")
print("="*60)

comparison_metrics = pd.DataFrame({
    'Dataset Split': ['Training Set', 'Validation Set', 'Unseen Test Set'],
    'MAE': [
        mean_absolute_error(y_train_true, y_pred_train),
        mean_absolute_error(y_val_true, y_pred_val),
        mean_absolute_error(y_test_true, y_pred_test)
    ],
    'R² Score': [
        r2_score(y_train_true, y_pred_train),
        r2_score(y_val_true, y_pred_val),
        r2_score(y_test_true, y_pred_test)
    ]
})

print(comparison_metrics.to_string(index=False))
print("="*60)
```
