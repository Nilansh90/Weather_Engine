## Converting datatypes
->so our csv has columns ->  #   Column             Non-Null Count  Dtype  
---  ------             --------------  -----  
 0   date               5476 non-null   str    
 1   city_id            5476 non-null   int64  
 2   weather_code       5476 non-null   int64  
 3   temp_max           5476 non-null   float64
 4   temp_min           5476 non-null   float64
 5   wind_speed_max     5476 non-null   float64
 6   wind_dir_dominant  5476 non-null   float64
 7   precip_sum         5476 non-null   float64
 8   precip_hours       5476 non-null   float64
 9   data_type          5476 non-null   str    
 
I have decided to->

1)Convert to date time format ,

2)convert weather code to a categorical colmn 

3)wind direction from degree to again 8 major labels
   
4)drop data type , 

there are non null values ,  and then do further eda 

```python
# First, define the helper function for wind direction
def deg_to_compass(degree):
    val = int((degree / 45) + 0.5)
    arr = ["N", "NE", "E", "SE", "S", "SW", "W", "NW"]
    return arr[(val % 8)]

# Next, create a function that applies all your steps to ANY dataframe you give it
def clean_weather_data(df):
    # 1) Convert to date time format
    df['date'] = pd.to_datetime(df['date'])
    
    # 2) Convert weather code to categorical (using your dictionary from Cell 26)
    df['weather_desc'] = df['weather_code'].map(weather_codes).astype('category')
    df['weather_code'] = df['weather_code'].astype('category')
    
    # 3) Wind direction from degree to 8 major labels
    df['wind_dir_label'] = df['wind_dir_dominant'].apply(deg_to_compass).astype('category')
    
    # 4) Drop data_type column (checking if it exists first so it doesn't error if run twice)
    if 'data_type' in df.columns:
        df.drop(columns=['data_type'], inplace=True)
        
    return df

# Finally, process your cities one by one!
DEL_data = clean_weather_data(DEL_data)
MUM_data = clean_weather_data(MUM_data)
NAG_data = clean_weather_data(NAG_data)
PIL_data = clean_weather_data(PIL_data)
BLRU_data = clean_weather_data(BLRU_data)

# Verify it worked on one of them
DEL_data.info()
```

## Visualising

```python
#Temp vs date plot
def plot_temperature(df, city_name):
    plt.figure(figsize=(15, 6))
    sns.lineplot(data=df, x='date', y='temp_max', label='Max Temp', color='crimson')
    sns.lineplot(data=df, x='date', y='temp_min', label='Min Temp', color='royalblue')
    plt.title(f'{city_name}: Daily Maximum and Minimum Temperatures')
    plt.xlabel('Date')
    plt.ylabel('Temperature (°C)')
    plt.legend()
    plt.show()
```

```python
#wind speed vs date plot
def plot_wind_speed(df, city_name):
    plt.figure(figsize=(15, 4))
    sns.lineplot(data=df, x='date', y='wind_speed_max', color='darkorange', alpha=0.7)
    plt.title(f'{city_name}: Maximum Daily Wind Speed')
    plt.xlabel('Date')
    plt.ylabel('Wind Speed')
    plt.show()
```

```python
#precipitation sum vs date
def plot_precipitation(df, city_name):
    plt.figure(figsize=(15, 4))
    sns.lineplot(data=df, x='date', y='precip_sum', color='teal')
    plt.title(f'{city_name}: Daily Precipitation Sum')
    plt.xlabel('Date')
    plt.ylabel('Precipitation')
    plt.show()
```

```python
#categorical plots
def plot_categorical_weather(df, city_name):
    fig, ax = plt.subplots(1, 2, figsize=(18, 6))
    
    # Wind Direction Count
    sns.countplot(data=df, x='wind_dir_label', order=['N', 'NE', 'E', 'SE', 'S', 'SW', 'W', 'NW'], palette='viridis', ax=ax[0])
    ax[0].set_title(f'{city_name}: Frequency of Wind Directions')
    
    # Weather Description Count
    sns.countplot(data=df, y='weather_desc', palette='Set2', ax=ax[1])
    ax[1].set_title(f'{city_name}: Frequency of Weather Conditions')
    
    plt.tight_layout()
    plt.show()
```

```python
#correlation plot
def plot_correlation(df, city_name):
    plt.figure(figsize=(10, 8))
    
    # Isolate only the numerical columns for the correlation matrix
    numeric_df = df.select_dtypes(include=['float64', 'int64'])
    corr_matrix = numeric_df.corr()
    
    # Plot the heatmap
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, vmin=-1, vmax=1)
    
    plt.title(f'{city_name}: Correlation Heatmap of Weather Variables')
    plt.tight_layout()
    plt.show()
```

## Splitting data into train val test splits and saving locally

##### self comment####
, now what we have done is eda and data improvement as changes mentioned above , now we should split data into train test , validation and then save those into processed data directory 

```python
def split_time_series_data(df, train_pct=0.70, val_pct=0.15):
    """Splits a dataframe sequentially into train, validation, and test sets."""
    
    # Calculate the exact row numbers for the splits
    train_end = int(len(df) * train_pct)
    val_end = int(len(df) * (train_pct + val_pct))
    
    # Slice the dataframe using .iloc (index location)
    train_set = df.iloc[:train_end]
    val_set = df.iloc[train_end:val_end]
    test_set = df.iloc[val_end:]
    
    print(f"Total rows: {len(df)}")
    print(f"Train: {len(train_set)} rows | Val: {len(val_set)} rows | Test: {len(test_set)} rows")
    
    return train_set, val_set, test_set

```

```python
##Splitting delhi data


DEL_train, DEL_val, DEL_test = split_time_series_data(DEL_data, train_pct=0.70, val_pct=0.15)
```

```python
##Saving the split data locally

import os

def save_split_data(train_df, val_df, test_df, city_name):
    """
    Creates the necessary folder structure and saves the split dataframes as CSVs.
    """
    # Define the path relative to your notebooks folder
    processed_dir = f"../data/processed/{city_name}"
    
    # Create the city folder inside data/processed/ (exist_ok=True prevents errors if it already exists)
    os.makedirs(processed_dir, exist_ok=True)
    
    # Save the splits to CSV
    train_df.to_csv(f"{processed_dir}/train.csv", index=False)
    val_df.to_csv(f"{processed_dir}/val.csv", index=False)
    test_df.to_csv(f"{processed_dir}/test.csv", index=False)
    
    print(f"✅ Saved splits for {city_name} in {processed_dir}/")

```

```python
#for Delhi 
save_split_data(DEL_train, DEL_val, DEL_test, "Delhi")
```

```python
# Create a dictionary of the remaining city dataframes
remaining_cities = {
    "Mumbai": MUM_data,
    "Nagpur": NAG_data,
    "Pilani": PIL_data,
    "Bangalore": BLRU_data
}

# Loop through and process each city automatically
for city_name, df in remaining_cities.items():
    print(f"--- Processing {city_name} ---")
    
    # 1. Split the data
    train_df, val_df, test_df = split_time_series_data(df, train_pct=0.70, val_pct=0.15)
    
    # 2. Save the data to the correct folder
    save_split_data(train_df, val_df, test_df, city_name)
    
    print("\n") # Adds a blank line for readability between cities
```

## Model gonna be used is light gbm , preparing it for lightgbm

```python
def prep_for_lightgbm(df):
    # 1. Extract numerical features from the date
    df['year'] = df['date'].dt.year
    df['month'] = df['date'].dt.month
    df['day'] = df['date'].dt.day
    df['dayofweek'] = df['date'].dt.dayofweek
    
    # Drop the original datetime column
    df.drop(columns=['date'], inplace=True)
    
    # 2. Drop redundant columns (cleaning up what was commented out earlier)
    redundant_cols = ['weather_code', 'wind_dir_dominant']
    # We use a list comprehension just to be safe in case they are already dropped
    df.drop(columns=[c for c in redundant_cols if c in df.columns], inplace=True)
    
    return df


```

```python
# List of all your cities
all_cities_list = ["Delhi", "Mumbai", "Nagpur", "Pilani", "Bangalore"]

for city in all_cities_list:
    print(f"⚙️ Preparing LightGBM data for {city}...")
    processed_dir = f"../data/processed/{city}"
    
    # Loop through the three split files for each city
    for split in ['train', 'val', 'test']:
        file_path = f"{processed_dir}/{split}.csv"
        
        # 1. Load the saved split
        df = pd.read_csv(file_path)
        
        # 2. Convert 'date' back to datetime so your function can extract the year/month/day
        df['date'] = pd.to_datetime(df['date'])
        
        # 3. Apply your preparation function
        df = prep_for_lightgbm(df)
        
        # 4. Overwrite the CSV with the newly prepared LightGBM data
        df.to_csv(file_path, index=False)
        
    print(f"✅ {city} splits successfully updated!")
```

## Data conversion as per LigthGBM

```python
DEL_data = pd.read_csv("../data/raw/Delhi_history.csv")
MUM_data = pd.read_csv("../data/raw/Mumbai_history.csv")
NAG_data = pd.read_csv("../data/raw/Nagpur_history.csv")
PIL_data = pd.read_csv("../data/raw/Pilani_history.csv")
BLRU_data=pd.read_csv("../data/raw/Bangalore_history.csv")
```

```python
def process_og_weather_data(df):
    """
    Processes original raw weather dataframes directly for LightGBM modeling,
    including cyclic month encoding and native categorical feature casting.
    """
    df = df.copy()
    
    # 1. Split Date into Features
    df['date'] = pd.to_datetime(df['date'])
    df['year'] = df['date'].dt.year
    df['month'] = df['date'].dt.month
    df['day'] = df['date'].dt.day
    df['dayofweek'] = df['date'].dt.dayofweek
    df = df.drop(columns=['date'])

    # 2. Add Cyclic Month Transformation (Keep as floats)
    df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
    df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)
    df = df.drop(columns=['month'])

    # 3. Bin Wind Direction into 8 Cardinal Labels (Mapped to 0-7)
    def bin_wind_direction(degrees):
        deg = degrees % 360
        bins = ['N', 'NE', 'E', 'SE', 'S', 'SW', 'W', 'NW']
        return bins[int((deg + 22.5) % 360 / 45)]

    if 'wind_dir_dominant' in df.columns:
        wind_num_map = {'N': 0, 'NE': 1, 'E': 2, 'SE': 3, 'S': 4, 'SW': 5, 'W': 6, 'NW': 7}
        df['wind_dir_label'] = df['wind_dir_dominant'].apply(bin_wind_direction).map(wind_num_map)
        df = df.drop(columns=['wind_dir_dominant'])

    # 4. Drop Redundant / Metadata Columns
    columns_to_drop = ['data_type', 'weather_desc']
    df = df.drop(columns=[col for col in columns_to_drop if col in df.columns])

    # 5. Convert Specified Features to LightGBM Categories
    cat_cols = ['city_id', 'weather_code', 'wind_dir_label', 'dayofweek']
    for col in cat_cols:
        if col in df.columns:
            df[col] = df[col].astype('category')
            
    return df
```

```python
# 1. Re-read the pure raw data files
raw_cities = {
    "Delhi": pd.read_csv("../data/raw/Delhi_history.csv"),
    "Mumbai": pd.read_csv("../data/raw/Mumbai_history.csv"),
    "Nagpur": pd.read_csv("../data/raw/Nagpur_history.csv"),
    "Pilani": pd.read_csv("../data/raw/Pilani_history.csv"),
    "Bangalore": pd.read_csv("../data/raw/Bangalore_history.csv")
}

# 2. Process, Split, and Overwrite saved CSVs
for city_name, raw_df in raw_cities.items():
    print(f"⚙️ Running engineering pipeline for {city_name}...")
    
    # Apply our new custom feature function
    processed_df = process_og_weather_data(raw_df)
    
    # Split the data sequentially (70% Train, 15% Val, 15% Test)
    train_df, val_df, test_df = split_time_series_data(processed_df, train_pct=0.70, val_pct=0.15)
    
    # Save/Overwrite to processed directory
    save_split_data(train_df, val_df, test_df, city_name)
    print("-" * 50)

# Quick Sanity Check on the newly saved data layout
print("\n Final Check on Processing Layout:")
pd.read_csv("../data/processed/Delhi/train.csv").info()
```
